```python
#Aggregated transaction
path="./pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
clm={'State':[],
     'Year':[],
     'Quarter':[],
     'Transaction_type':[],
     'transaction_count':[],
     'transaction_amount':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i) #lists aggregated year 
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r') #opening the file in the read only mode
            d=json.load(Data)
            for z in d['data']['transactionData']:#Getting into transaction list
                Name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                #appending details into the clm list
                clm['Transaction_type'].append(Name)
                clm['transaction_count'].append(count)
                clm['transaction_amount'].append(amount)
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quarter'].append(int(k.strip('.json')))#removes .json 
#dataframe
Agg_Trans=pd.DataFrame(clm)

    
```

```python
#Top Transaction
path="./pulse/data/top/transaction/country/india/state/"
state_list = os.listdir(path)
clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Level': [],             # State / District / Pincode
    'Metric_type': [],       # Name like Karnataka or Bengaluru Urban
    'Metric_count': [],
    'Metric_amount': []
}

for state in state_list:
    state_path = os.path.join(path, state)
    years = os.listdir(state_path)
    
    for year in years:
        year_path = os.path.join(state_path, year)
        quarters = os.listdir(year_path)
        
        for quarter_file in quarters:
            file_path = os.path.join(year_path, quarter_file)
            quarter = int(quarter_file.strip('.json'))

            with open(file_path, 'r') as f:
                data = json.load(f)

            data_section = data.get('data', {})

            # Extract from each level: states, districts, pincodes
            for level_name in ['states', 'districts', 'pincodes']:
                items = data_section.get(level_name) or []
                for item in items:
                    clm['State'].append(state)
                    clm['Year'].append(year)
                    clm['Quarter'].append(quarter)
                    clm['Level'].append(level_name[:-1].capitalize())  # singular: State/District/Pincode
                    clm['Metric_type'].append(item.get('entityName', 'Unknown'))

                    metric = item.get('metric', {})
                    clm['Metric_count'].append(metric.get('count', 0))
                    clm['Metric_amount'].append(metric.get('amount', 0.0))

# Convert to DataFrame
Top_Trans = pd.DataFrame(clm)
```

```python
import os
import json
import pandas as pd

# Path to your user data
path = "./pulse/data/top/user/country/india/state/"

# Check if the path exists
if not os.path.exists(path):
    raise FileNotFoundError(f"The path {path} does not exist!")

state_list = os.listdir(path)
clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Level': [],
    'Metric_type': [],
    'RegisteredUsers': []
}

# Loop through each state folder
for state in state_list:
    state_path = os.path.join(path, state)
    if not os.path.isdir(state_path):
        continue  # Skip if it's not a directory
    
    years = os.listdir(state_path)
    
    for year in years:
        year_path = os.path.join(state_path, year)
        if not os.path.isdir(year_path):
            continue
        
        quarters = os.listdir(year_path)
        
        for quarter_file in quarters:
            file_path = os.path.join(year_path, quarter_file)
            try:
                quarter = int(quarter_file.strip('.json'))
            except ValueError:
                continue  # Skip if the quarter filename is not in the expected format

            try:
                with open(file_path, 'r') as f:
                    data = json.load(f)
            except (IOError, json.JSONDecodeError):
                continue  # Skip files that can't be read or decoded
            
            data_section = data.get('data', {})
            
            # Loop through each level
            for level_name in ['states', 'districts', 'pincodes']:
                items = data_section.get(level_name) or []
                for item in items:
                    clm['State'].append(state)
                    clm['Year'].append(year)
                    clm['Quarter'].append(quarter)
                    clm['Level'].append(level_name[:-1].capitalize())  # Convert plural to singular
                    clm['Metric_type'].append(item.get('name', 'Unknown'))
                    clm['RegisteredUsers'].append(item.get('registeredUsers', 0))

# Convert to DataFrame
Top_Users = pd.DataFrame(clm)
```

```python
import json
import csv
from pathlib import Path

# Define the notebook file path (you can change this accordingly)
notebook_path = Path("PHONEPE (1).ipynb")
output_csv_path = notebook_path.with_name("PHONEPE_outputs_final.csv")

# Load the .ipynb file
with open(notebook_path, "r", encoding="utf-8") as f:
    notebook = json.load(f)

# Extract only the outputs
outputs_only = []

for cell in notebook.get("cells", []):
    if cell.get("cell_type") == "code":
        for output in cell.get("outputs", []):
            if output.get("output_type") == "stream":
                outputs_only.append(["".join(output.get("text", []))])
            elif output.get("output_type") == "execute_result":
                data = output.get("data", {})
                outputs_only.append(["".join(data.get("text/plain", []))])
            elif output.get("output_type") == "error":
                outputs_only.append(["\n".join(output.get("traceback", []))])

# Save outputs to CSV
with open(output_csv_path, "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["output"])
    writer.writerows(outputs_only)
```

```python
import sqlite3
import pandas as pd

# Connect using sqlite3
conn = sqlite3.connect('phonepe_data.db')
cursor = conn.cursor()

# Get table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cursor.fetchall()]

# Export each table
for table in tables:
    df = pd.read_sql_query(f"SELECT * FROM {table}", conn)
    df.to_csv(f"{table}.csv", index=False)
    print(f"Exported {table}.csv")

conn.close()
```
